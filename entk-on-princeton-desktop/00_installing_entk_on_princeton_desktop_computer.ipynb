# 0: EnTK on Princeton University Desktop Computer

## 0.1: Before installing EnTK:

**1. Check that desktop is running Springdale Linux 7.\*:**

- Open up a terminal window on your desktop and execute the following command

```bash
cat /etc/redhat-release
```

**2. Check that Python distribution is Python 2.7.\*:**

- Open up a terminal window on your desktop and type

```bash
python --version
```

**3. Check that Pip is installed on your machine:**

`pip` is package manager for Python that we will use to install EnTK and it's dependencies.

- Open a terminal window on your desktop and type

```bash
pip --version
```

- If the command exists, make sure that it says `(python 2.7)`. If it does not exist, you need to ask the system administrator to install it for you.

**4. Install `virtualenv`:**

If `virtualenv` is not already available on your desktop, it can be installed using `pip` as follows:

- Open up a terminal window on your desktop and run the command

```bash
pip install --user virtualenv
```

- This will install `virtualenv` for your user, but it might need to be manually added to your system path. This is done as follows:

```bash
echo PATH=~/.local/bin:$PATH >> ~/.bash_profile && source ~/.bash_profile
```

- Finally, run the following command to check that `virtualenv` is now available

```bash
virtualenv --version
```

If you get a version number, you are good to go.

**5. Install `rabbitmq`:**

[RabbitMQ](https://www.rabbitmq.com/) is an application used by EnTK for message transfer between its internal components. It requires root access to install, and must therefore be installed by a system administrator.

## 0.2: Installing EnTK

The following steps will show you one way of installing EnTK on your desktop. For a more elaborate discussion, visit the [official installation page](https://radicalentk.readthedocs.io/en/latest/install.html).

With the above steps taken care of, we can install EnTK in a virtual environment as follows:

**1. Set up a new virtual environment:**

- Open a terminal on your desktop and run the following command to create a new virtual environment in `$HOME`

```bash
virtualenv $HOME/venv
```

- Activate the virtual environment:

```bash
source $HOME/venv/bin/activate
```

- Check that the virtual environment is active by typing

```bash
which python
```

And making sure that it returns: `$HOME/venv/bin/python`.

**2. Install EnTK using `pip`:**

- Install EnTK by running the following command in your terminal:

```bash
pip install radical.entk
```

- If you prefer to install a specific version of EnTK instead, this can be done with the following command instead of the above command:

```bash
pip install radical.entk==<version number>
```

## 0.3: Running EnTK

To test the EnTK installation, we will run a "Hello World!" example script on the local machine. Complete instructions and additional example scripts can be found in the official [user guide](https://radicalentk.readthedocs.io/en/latest/user_guide.html). All the example scripts are located in the virtual environment in the folder

```bash
$HOME/venv/share/radical.entk/user_guide/scripts
```

Before running the example script, we need to set two environment variables, which are `RADICAL_PILOT_VERBOSE` and `RADICAL_PILOT_DBURL`. The former sets the verbose option for the execution, and the latter sets the MongoDB database url to be used during execution. Set these by executing these two commands in the terminal:

```bash
export RADICAL_PILOT_VERBOSE="REPORT"
export RADICAL_PILOT_DBURL="mongodb://user:user123@ds043012.mlab.com:43012/princeton"
```

It can be convenient to set these environment variables in your `.bash_profile` if you plan to run EnTK on a regular basis.

With the environment variables taken care of, we shift focus to obtaining the example script. We want to run the `get_started.py` script from the examples folder, so we first copy it to `$HOME`:

```bash
cp $HOME/venv/share/radical.entk/user_guide/scripts/get_started.py $HOME
```

Before running the script, make sure that the virtual environment is activated:

```bash
which python
# ~/venv/bin/python
```

Then run the script:

```bash
cd $HOME && python get_started.py
```
